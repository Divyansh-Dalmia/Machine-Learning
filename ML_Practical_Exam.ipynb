```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram
```

```python
# Generate a sample dataset
n_samples = 300
X, _ = make_blobs(n_samples=n_samples, centers=4, random_state=42)
```

```python
# Perform Agglomerative Clustering
n_clusters = 3  # You can choose the number of clusters
agg_clustering = AgglomerativeClustering(n_clusters=n_clusters)
agg_clustering.fit(X)
```

```python
# Plot the clustering result
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=agg_clustering.labels_, cmap='viridis')
plt.title('Agglomerative Clustering')
plt.show()
```

```python
# Plot the dendrogram (optional)
from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
linkage_matrix = linkage(X, method='ward')  # You can change the linkage method
dendrogram(linkage_matrix)
plt.title('Dendrogram')
plt.show()
```

We import the necessary libraries.
We generate a synthetic dataset using make_blobs. You can replace this with your own dataset.
We create an instance of AgglomerativeClustering with the desired number of clusters and fit it to the data.
We visualize the clustering results using a scatter plot.
Optionally, we can also plot the dendrogram to visualize the hierarchical clustering structure.

Agglomerative clustering is a hierarchical clustering technique that merges data points or clusters into larger clusters iteratively. You can perform agglomerative clustering in Python using the scikit-learn library, and I'll provide you with a code example that you can run in Google Colab.

Kmeans Clustring 2 Niguel

```python
from sklearn.datasets import load_breast_cancer
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# Load the breast cancer dataset from sklearn
data = load_breast_cancer()
```

```python
# Considering the first five features for clustering
X = data.data[:, :5]
```

```python
# Initialize the KMeans model
kmeans = KMeans(n_clusters=5, random_state=42)
```

```python
# Fit the model with the dataset
kmeans.fit(X)
```

```python
# Obtain cluster labels and cluster centers
labels = kmeans.labels_
centers = kmeans.cluster_centers_
```

```python
# Visualize the clustered data points and cluster centers
plt.figure(figsize=(8, 6))
```

```python
# Plotting data points with their assigned clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50, alpha=0.5, label='Data Points')

# Plotting cluster centers
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.9, label='Cluster Centers')

plt.title('K-Means Clustering (Using 5 Features)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
```

```python
# Load the iris dataset
iris = load_iris()
X = iris.data  # Features
```

```python
# Create a K-Means clustering model with 3 clusters (since there are 3 classes in the iris dataset)
kmeans = KMeans(n_clusters=3, random_state=42)
```

```python
# Fit the model to the data
kmeans.fit(X)
```

```python
# Get cluster labels and cluster centers
labels = kmeans.labels_
centers = kmeans.cluster_centers_
```

```python
# Plot the data points with cluster assignments
plt.figure(figsize=(8, 6))
```

```python
# Plot data points
plt.scatter(X[:, 0], X[:, 1], c=labels, edgecolors='k', marker='o', s=50, linewidth=1, cmap=plt.cm.Paired)
```

```python
# Plot the data points with cluster assignments
plt.figure(figsize=(8, 6))

# Plot data points
plt.scatter(X[:, 0], X[:, 1], c=labels, edgecolors='k', marker='o', s=50, linewidth=1, cmap=plt.cm.Paired)

# Plot cluster centers
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='X', s=200, linewidth=1, alpha=0.7)

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-Means Clustering')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('/content/drive/My Drive/student_clustering.csv')  # Replace 'your_data.csv' with your dataset file path
columns_of_interest = ['cgpa', 'iq']
subset = data[columns_of_interest]

# Calculate the IQR (Interquartile Range) for each column
Q1 = subset.quantile(0.25)
Q3 = subset.quantile(0.75)
IQR = Q3 - Q1
outlier_threshold = 1.5
# Find outliers using the IQR method
outliers = ((subset < (Q1 - outlier_threshold * IQR)) | (subset > (Q3 + outlier_threshold * IQR))).any(axis=1)
plt.figure(figsize=(10, 6))
plt.scatter(data['cgpa'], data['iq'], label='Data Points')

plt.scatter(data[outliers]['cgpa'], data[outliers]['iq'], color='red', label='Outliers')

plt.xlabel('CGPA')
plt.ylabel('IQ')
plt.title('Scatter Plot of cgpa vs. iq with Outliers')
plt.legend()
plt.show()
```
